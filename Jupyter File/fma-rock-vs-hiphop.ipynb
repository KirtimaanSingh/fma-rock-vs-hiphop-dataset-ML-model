```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read both dataset
tracks = pd.read_csv(r'../datasets/fma-rock-vs-hiphop.csv')

echonest_metrics = pd.read_json(r'../datasets/echonest-metrics.json', precise_float=True)

# Combining tracks and echonest_metrics relevant columns
echotracks = echonest_metrics.merge(tracks[['track_id', 'genre_top']], on='track_id')
```

```python
cmetrics = echotracks.corr()
cmetrics.style.background_gradient(cmap='Greens')
```

```python
mainf = echotracks.drop(['genre_top','track_id'],axis=1)
labels = echotracks['genre_top']
```

```python
# Import the StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_train_mainf =scaler.fit_transform(mainf)
```

```python
# Import train_test_split function and Decision tree classifier and predict the labels of test cases.
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
train_mainf, test_mainf, train_labels, test_labels = train_test_split(scaled_train_mainf, labels, random_state=10)
tree = DecisionTreeClassifier(random_state=10)
tree.fit(train_mainf, train_labels)
pred_labels_tree = tree.predict(test_mainf)
```

```python
# Import Logistic regression and Train and predict labels for test sets with Logistic regression.
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=10)
logreg.fit(train_mainf, train_labels)
pred_labels_logit = logreg.predict(test_mainf)
```

```python
# Print the reports
from sklearn.metrics import classification_report
class_rep_tree = classification_report(test_labels, pred_labels_tree)
class_rep_log = classification_report(test_labels, pred_labels_logit)

print("Decision Tree: \n", class_rep_tree)
print("Logistic Regression: \n", class_rep_log)
```

```python
# To get different result,train with same number of rock and hiphop songs
onlyhiphop = echotracks.loc[echotracks["genre_top"] == "Hip-Hop"]
onlyrock = echotracks.loc[echotracks["genre_top"] == "Rock"].sample(len(onlyhiphop), random_state=10)
rockhiphopbal = pd.concat([onlyrock, onlyhiphop])
from sklearn.decomposition import PCA
pca = PCA()
mainf = rockhiphopbal.drop(['genre_top', 'track_id'], axis=1) 
labels = rockhiphopbal['genre_top']
scaled_train_mainf = pca.fit_transform(scaler.fit_transform(mainf))

train_mainf, test_mainf, train_labels, test_labels = train_test_split(scaled_train_mainf, labels, random_state=10)
```

```python

#Decision Tree
tree = DecisionTreeClassifier(random_state=10)
tree.fit(train_mainf, train_labels)
pred_labels_tree = tree.predict(test_mainf)

#Logistic regression 
logreg = LogisticRegression(random_state=10)
logreg.fit(train_mainf, train_labels)
pred_labels_logit = logreg.predict(test_mainf)

# Print reports with balanced sets
print("Decision Tree: \n", classification_report(test_labels, pred_labels_tree))
print("Logistic Regression: \n", classification_report(test_labels, pred_labels_logit))
```

```python
#Cross-Validatiom
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=10, shuffle = True, random_state=10)

tree = DecisionTreeClassifier(random_state=10)
logreg = LogisticRegression(random_state=10)

dtreeval = cross_val_score(tree, scaled_train_mainf, labels, cv=kf)
logrval = cross_val_score(logreg, scaled_train_mainf, labels, cv=kf)
a = np.mean(dtreeval)
b = np.mean(logrval)
# a = dtreeval
# b = logrval

print("Decision Tree:", a, "Logistic Regression:", b)
```

Logistic Regression is more accurate.
